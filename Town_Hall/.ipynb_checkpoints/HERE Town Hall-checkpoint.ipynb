```python
vac = pd.read_excel("vac_covid.xlsx")
vac
```

```python
citiesVac = vac[vac['Table 1'] == 'City of Calabasas']
citiesVac = citiesVac.append(vac[vac['Table 1'] == 'City of Malibu'])
citiesVac = citiesVac.append(vac[vac['Table 1'] == "Los Angeles - Pacoima"])
citiesVac = citiesVac.append(vac[vac['Table 1'] == "Los Angeles - Canoga Park"])
citiesVac = citiesVac.append(vac[vac['Table 1'] == "Los Angeles - Northridge"])
citiesVac = citiesVac.append(vac[vac['Table 1'] == "Los Angeles - Panorama City"])
citiesVac = citiesVac.append(vac[vac['Table 1'] == "Los Angeles - Reseda"])
citiesVac = citiesVac.append(vac[vac['Table 1'] == "Los Angeles - Van Nuys"])
citiesVac.columns = ['geo_merge', "Number of Persons Vaccinated", "Population 2019", "Percentage of People Vaccinated"]
citiesVac
```

```python


### We wil use 

N = 8

# Data on X-axis

# Specify the values of blue bars (height)
blue_bar = citiesVac['Number of Persons Vaccinated']
# Specify the values of orange bars (height)
orange_bar = citiesVac['Population 2019']

# Position of bars on x-axis
ind = np.arange(N)

# Figure size
plt.figure(figsize=(10,5))

# Width of a bar 
width = 0.35       

# Plotting
ax = plt.bar(ind, blue_bar , width, label='Number of Persons Vaccinated', color = 'r')
ax = plt.bar(ind + width, orange_bar, width, label='Population 2019', color = 'y')

plt.xlabel('Los Angeles Cities', fontsize = 15)
plt.ylabel('Number of Population', fontsize = 15)
plt.title('Percentage of Persons Vaccinated', fontsize = 20)

# xticks()
# First argument - A list of positions at which ticks should be placed
# Second argument -  A list of labels to place at the given locations
plt.xticks(ind + width / 2, citiesVac['geo_merge'])


# Finding the best position for legends and putting it
plt.legend(loc='best', prop={'size': 15})
plt.xticks(rotation = 50)
plt.show()
```

```python
race = pd.read_excel('here_race.xlsx')
race = race.rename(columns=race.iloc[0]).drop(race.index[0])
race
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
import pandas as pd
 
# Data
r = [0,1,2,3,4,5,6,7]

df = race.drop(columns = ['City'])
 
# From raw value to percentage
totals = [i+j+k for i,j,k,a,b,c in zip(df['Hispanic'], df['White'], df['African American'], df['Asian'], df['Two or more races'],df['Other race'])]
greenBars = [i / j * 100 for i,j in zip(df['Hispanic'], totals)]
orangeBars = [i / j * 100 for i,j in zip(df['White'], totals)]
blueBars = [i / j * 100 for i,j in zip(df['African American'], totals)]
purpleBars = [i / j * 100 for i,j in zip(df['Asian'], totals)]
yellowBars = [i / j * 100 for i,j in zip(df['Two or more races'], totals)]
redBars = [i / j * 100 for i,j in zip(df['Other race'], totals)]
plt.figure(figsize=(10,15))
# plot
barWidth = 0.85
names = race['City']
# Create green Bars
plt.bar(r, greenBars, color='#b5ffb9', edgecolor='white', width=barWidth)
# Create orange Bars
plt.bar(r, orangeBars, bottom=greenBars, color='#f9bc86', edgecolor='white', width=barWidth)
# Create blue Bars
plt.bar(r, blueBars, bottom=[i+j for i,j in zip(greenBars, orangeBars)], color='#a3acff', edgecolor='white', width=barWidth)

plt.bar(r, purpleBars, bottom=[i+j for i,j,k in zip(greenBars, orangeBars,blueBars)], color='purple', edgecolor='white', width=barWidth)

plt.bar(r, yellowBars, bottom=[i+j for i,j,k,a in zip(greenBars, orangeBars,blueBars, purpleBars)], color='yellow', edgecolor='white', width=barWidth)

plt.bar(r, redBars, bottom=[i+j for i,j,k,a,b in zip(greenBars, orangeBars,blueBars, purpleBars,yellowBars)], color='red', edgecolor='white', width=barWidth)

 
# Custom x axis
plt.xticks(r, names)
plt.xlabel("group")
 
# Show graphic
plt.show()
```

```python
import matplotlib.pyplot as plt


labels = race['City']

width = 0.5       # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots()
fig.set_size_inches(10,5)
ax.bar(labels, race['Hispanic'], width,  label='Hispanic')
ax.bar(labels, race['White'], width, bottom= race['Hispanic'], label='White')
ax.bar(labels, race['African American'], width, bottom= race['White'] + race['Hispanic'], label='African American')
ax.bar(labels, race['Asian'], width, bottom= race['White'] + race['Hispanic'] + race['African American'], label='Asian')
ax.bar(labels, race['Two or more races'], width, bottom= race['White'] + race['Hispanic'] + race['African American'] + race['Asian'], label='Two or More Races')
ax.bar(labels, race['Other race'], width, bottom= race['White'] + race['Hispanic'] + race['African American'] + race['Asian'] + race['Two or more races'], label='Other Race')

ax.set_ylabel('Percentage', fontsize = 15)
ax.set_xlabel('Los Angeles Cities', fontsize = 15)
ax.set_title('Demographics by Cities', fontsize = 20)
leg = ax.legend( prop={'size': 13},loc='lower left')
leg.get_frame().set_alpha(.8)
vals = ax.get_yticks()
ax.set_yticklabels(['{:,.0%}'.format(x) for x in vals])
plt.xticks(rotation = 50)

plt.show()
```
